```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv("updated_records.csv")
df = df.drop(columns=["Rk","Conf", "Coach(es)"])
df.reset_index(drop=True)
for i in range(len(df)):
   if pd.isna(df.iloc[i]["NCAA Tournament"]):
      df.loc[df.index[i], "NCAA Tournament"] = "Missed Tournament"

   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament First Round":
      df.loc[df.index[i], "NCAA Tournament"] = "First Round"

   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Second Round" or df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Third Round":
      df.loc[df.index[i], "NCAA Tournament"] = "Second Round"

   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Regional Semifinal":
      df.loc[df.index[i], "NCAA Tournament"] = "Sweet Sixteen"

   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Regional Final":
      df.loc[df.index[i], "NCAA Tournament"] = "Elite Eight"
   
   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament National Semifinal":
      df.loc[df.index[i], "NCAA Tournament"] = "Final Four"
   
   elif df.loc[df.index[i], "NCAA Tournament"] == "Lost NCAA Tournament National Final":
      df.loc[df.index[i], "NCAA Tournament"] = "National Runner-Up"
   
   elif df.loc[df.index[i], "NCAA Tournament"] == "Won NCAA Tournament National Final":
      df.loc[df.index[i], "NCAA Tournament"] = "National Champions"


df["Conf W"] = df["Conf W"].astype(int)
df["Conf L"] = df["Conf L"].astype(int)

df["Non-Conf W"] = df["Overall W"] - df["Conf W"]
df["Non-Conf L"] = df["Overall L"] - df["Conf L"]

df.head()
```
