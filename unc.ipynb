```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("unc_records.csv",)
df.head(15)
```

```python
df.isna().sum()
```

```python
ncaa_64 = df[:42]
ncaa_64.isna().sum()
```

```python
ncaa_64 = ncaa_64[1:]
ncaa_64.head()
```

```python
ncaa_64 = ncaa_64.drop(columns=["Rk","Conf", "Coach(es)"])
ncaa_64.head()
```

```python
ncaa_64.reset_index(drop=True)
```

```python
for i in range(len(ncaa_64)):
    # print(i,"---------")
    # print(ncaa_64.iloc[i]["NCAA Tournament"])
   if pd.isna(ncaa_64.iloc[i]["NCAA Tournament"]):
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "Missed Tournament"

   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament First Round":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "First Round"

   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Second Round" or ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Third Round":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "Second Round"

   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Regional Semifinal":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "Sweet Sixteen"

   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament Regional Final":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "Elite Eight"
   
   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament National Semifinal":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "Final Four"
   
   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Lost NCAA Tournament National Final":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "National Runner-Up"
   
   elif ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] == "Won NCAA Tournament National Final":
      ncaa_64.loc[ncaa_64.index[i], "NCAA Tournament"] = "National Champions"
```

```python
ncaa_64[" Conf W"] = ncaa_64[" Conf W"].astype(int)
ncaa_64[" Conf L"] = ncaa_64[" Conf L"].astype(int)
```

```python
ncaa_64["NCAA Tournament"].value_counts()
```

```python
ax = ncaa_64["NCAA Tournament"].value_counts().sort_index().plot(
    kind="bar",
    figsize=(10, 5),
    rot=30
)

ax.set_xlabel("Tournament Round")
ax.set_ylabel("Count")
ax.set_title("NCAA Tournament Results")
```
